```python
#Smart Parking Lot Management System
def smart_parking(capacity, vehicle_logs):
    current_vehicles = 0
    peak_usage = 0

    for log in vehicle_logs:
        if log == "IN":
            current_vehicles += 1
        elif log == "OUT" and current_vehicles > 0:
            current_vehicles -= 1
        if current_vehicles > peak_usage:
            peak_usage = current_vehicles

    status = "Full" if current_vehicles >= capacity else "Available"

    print(f"Currently Parked Vehicles: {current_vehicles}")
    print(f"Parking Status: {status}")
    print(f"Peak Parking Usage: {peak_usage}")

smart_parking(50, ["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"])
```

```python
#Online Food Delivery Time Estimator
def delivery_time_estimator(distance_km, traffic, weather):
    base_time = distance_km * 5
    if traffic.lower() == "high":
        base_time += 15
    elif traffic.lower() == "medium":
        base_time += 5

    if weather.lower() == "rainy":
        base_time += 10
    elif weather.lower() == "stormy":
        base_time += 20

    print(f"Estimated Delivery Time: {base_time} minutes")

delivery_time_estimator(8, "High", "Rainy")
```

```python
#Movie Theatre Seat Occupancy Analyzer
def seat_occupancy(total_seats, booked_seats):
    booked_count = len(booked_seats)
    occupancy_percentage = (booked_count / total_seats) * 100

    if occupancy_percentage >= 100:
        status = "Housefull"
    elif occupancy_percentage >= 75:
        status = "Almost Full"
    else:
        status = "Available"

    print(f"Occupancy: {occupancy_percentage:.0f}%")
    print(f"Show Status: {status}")
seat_occupancy(200, [1]*150)
```

```python
#Cloud Server Load Classification System
def server_load_classification(cpu_readings):
    avg_cpu = sum(cpu_readings) / len(cpu_readings)

    if avg_cpu < 50:
        status = "Normal"
    elif avg_cpu < 80:
        status = "Warning"
    else:
        status = "Critical"

    print(f"Average CPU Load: {avg_cpu:.0f}%")
    print(f"Server Status: {status}")

server_load_classification([45, 60, 70, 85, 90])
```

```python
#Smart Classroom Resource Usage Monitor
def classroom_resource_monitor(resource_usage):
    overused = [resource for resource, hours in resource_usage.items() if hours > 8]

    print("Overused Resources:", ", ".join(overused) if overused else "None")
    print("Energy Alert:", "Yes" if overused else "No")

classroom_resource_monitor({
    "Projector": 6,
    "AC": 9,
    "Lights": 4
})
```

```python
# Online Event Registration Capacity Controller
def event_registration_controller(capacity, registrations):
    confirmed = min(capacity, registrations)
    waitlisted = max(0, registrations - capacity)
    status = "Closed" if registrations > capacity else "Open"

    print(f"Confirmed Registrations: {confirmed}")
    print(f"Waitlisted Users: {waitlisted}")
    print(f"Registration Status: {status}")

event_registration_controller(100, 105)
```

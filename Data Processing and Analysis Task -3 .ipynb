```python
# Search Query Keyword Analysis
search_query = "Buy mobile phone buy phone online"
search_query = search_query.lower()
cleaned_query = ""
for character in search_query:
    if character.isalnum() or character.isspace():
        cleaned_query += character
words_list = cleaned_query.split()
keyword_count = {}
for word in words_list:
    if word in keyword_count:
        keyword_count[word] += 1
    else:
        keyword_count[word] = 1
repeated_keywords = {}

for word, count in keyword_count.items():
    if count > 1:
        repeated_keywords[word] = count
print(repeated_keywords)
```

```python
# Sensor Data Validation
sensor_readings = [3, 4, 7, 8, 10, 12, 5]
valid_readings = []
for hour in range(len(sensor_readings)):
    reading = sensor_readings[hour]
    if reading % 2 == 0:
        valid_readings.append((hour, reading))
print("Valid Sensor Readings (Hour, Value):")
print(valid_readings)
```

```python
# Email Domain Usage Analysis
emails = [
    "ravi@gmail.com",
    "anita@yahoo.com",
    "kiran@gmail.com",
    "suresh@gmail.com",
    "meena@yahoo.com"
]
domain_count = {}
total_emails = len(emails)
for email in emails:
    domain = email.split("@")[1]

    if domain in domain_count:
        domain_count[domain] += 1
    else:
        domain_count[domain] = 1
for domain, count in domain_count.items():
    percentage = (count / total_emails) * 100
    print(f"{domain}: {int(percentage)}%")
```

```python
# Sales Spike Detection
sales = [1200, 1500, 900, 2200, 1400, 3000]
average_sales = sum(sales) / len(sales)
threshold = average_sales * 1.30
rounded_threshold = round(threshold, -2)
for day, value in enumerate(sales):
    if value >= rounded_threshold:
        print(f"Day {day + 1}: {value}")
```

```python
# Duplicate User ID Detection
user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]
id_count = {}
for user in user_ids:
    if user in id_count:
        id_count[user] += 1
    else:
        id_count[user] = 1
for user, count in id_count.items():
    if count > 1:
        print(f"{user} → {count} times")
```
